```python
df = pd.read_csv('Exercices_Pandas_athlete_events.csv')
# Step 1: Filter only rows with medals
df = df[df['Medal'].notna()]

# Step 2: Count medals per Team within each Sport
medal_counts = df.groupby(['Sport', 'NOC']).size().reset_index(name='Medal_Count')

# Step 3: For each Sport, find the Team with the most medals
top_teams = medal_counts.loc[medal_counts.groupby('Sport')['Medal_Count'].idxmax()]

top_teams.head(30)
```

```python
PT = df.pivot_table(index=['NOC','Sport'],
                      columns=['Medal'], 
                      values='ID',
                      margins=True, 
                      aggfunc='count'
                     ).sort_values(by=['All'],ascending=False)
PT.reset_index().drop_duplicates('Sport')

```

```python
df = pd.read_csv('Exercices_Pandas_athlete_events.csv')
cond = (df["Year"].between(1952,1988))
df_cold_war_person_medal_by_country = df[cond].groupby(['NOC']).agg({'Name':['count'],'Medal':['count']})
df_cold_war_person_medal_by_country['Efficiency'] = (df_cold_war_person_medal_by_country['Medal'] / df_cold_war_person_medal_by_country['Name'])*100 
display(df_cold_war_person_medal_by_country.nlargest(50,columns='Efficiency'))

```
